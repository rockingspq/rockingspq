```python
import pandas as pd  
import matplotlib.pyplot as plt    
import seaborn as sns        
%matplotlib inline
```

```python
df=pd.read_csv('Mcdonald .csv')
df.head(2)
```

```python
# varities by category
# coffee & Tea has highest varities & Salads has lowest.
```

```python
plt.figure(figsize=(10,8))
sns.countplot(y=df['Category'],palette='flare')

plt.show()
```

```python
# using boxplot to find the outlier in each varibles.
# all variables are having outlier except Saturated fat,Saturated Fat (% Daily Value) & Dietary Fiber
```

```python
fig, axes = plt.subplots(3, 7,figsize=(20,18))
sns.boxplot(ax=axes[0,0], data=df, x='Calories')
sns.boxplot(ax=axes[0,1], data=df, x='Calories from Fat')
sns.boxplot(ax=axes[0,2], data=df, x='Total Fat')
sns.boxplot(ax=axes[0,3], data=df, x='Total Fat (% Daily Value)')
sns.boxplot(ax=axes[0,4], data=df, x='Saturated Fat')
sns.boxplot(ax=axes[0,5], data=df, x='Saturated Fat (% Daily Value)')
sns.boxplot(ax=axes[0,6], data=df, x='Trans Fat')
sns.boxplot(ax=axes[1,0], data=df, x='Cholesterol')
sns.boxplot(ax=axes[1,1], data=df, x='Cholesterol (% Daily Value)')
sns.boxplot(ax=axes[1,2], data=df, x='Sodium')
sns.boxplot(ax=axes[1,3], data=df, x='Sodium (% Daily Value)')
sns.boxplot(ax=axes[1,4], data=df, x='Carbohydrates')
sns.boxplot(ax=axes[1,5], data=df, x='Carbohydrates (% Daily Value)')
sns.boxplot(ax=axes[1,6], data=df, x='Dietary Fiber')
sns.boxplot(ax=axes[2,0], data=df, x='Dietary Fiber (% Daily Value)')
sns.boxplot(ax=axes[2,1], data=df, x='Sugars')
sns.boxplot(ax=axes[2,2], data=df, x='Protein')
sns.boxplot(ax=axes[2,3], data=df, x='Vitamin A (% Daily Value)')
sns.boxplot(ax=axes[2,4], data=df, x='Vitamin C (% Daily Value)')
sns.boxplot(ax=axes[2,5], data=df, x='Calcium (% Daily Value)')
sns.boxplot(ax=axes[2,6], data=df, x='Iron (% Daily Value)')
plt.show()


```

```python
# finding highest correlation variables by plotting heatmap.Yellowish blocks showing the strongest correlation.
```

```python
plt.figure(figsize=(15,16))

sns.heatmap(df.corr(),annot=True,cmap='viridis');
```

```python
# category contributes to the maximum % of Cholesterol in a diet (% daily value)
# Breakfast category has highest cholesterol in a diet.
```

```python
ch=df[['Category','Cholesterol (% Daily Value)']].sort_values(by='Cholesterol (% Daily Value)',ascending=False).head(10)
plt.figure(figsize=(10,8))
sns.barplot(y='Category',x='Cholesterol (% Daily Value)',data=ch,palette='husl')
plt.show()
```

```python
# item contributes maximum to the Sodium intake
# maximum sodium intake is from Chicken McNuggets
```

```python
sf=df[['Item','Sodium']].sort_values(by='Sodium',ascending=False).head(10)
plt.figure(figsize=(10,8))
sns.barplot(y='Item',x='Sodium',data=sf,palette='rocket_r')
plt.show()
```

```python
# 4 food items contain the most amount of Saturated Fat
```

```python
SF=df[['Item','Saturated Fat']].sort_values(by='Saturated Fat',ascending=False).head(4)
sns.countplot(data=SF,y="Item");
```
